# **Importación de datos**



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.tail()
```

#**1. Análisis de facturación**
#### Calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.



```python
# Calcular el ingreso total para cada tienda
ingreso_tienda1 = tienda['Precio'].sum() # Cambiar tienda1 a tienda
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("\n" + "="*50)
print("ANÁLISIS DE INGRESOS TOTALES POR TIENDA".center(50))
print("="*50)
print(f"\n{'Tienda':<10} {'Ingreso Total':>20} {'Porcentaje':>15}")
print("-"*45)

# Calcular el ingreso combinado para el porcentaje
total_general = ingreso_tienda1 + ingreso_tienda2 + ingreso_tienda3 + ingreso_tienda4

for i, ingreso in enumerate([ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4], 1):
    porcentaje = (ingreso / total_general) * 100
    print(f"{f'Tienda {i}':<10} ${ingreso:>15,.2f} {porcentaje:>14.1f}%")

print("\n" + "-"*45)
print(f"{'TOTAL GENERAL:':<10} ${total_general:>15,.2f} {100:>14.1f}%")
print("="*50 + "\n")

# Versión alternativa compacta para resultados rápidos
print("Resumen rápido:".upper())
print(f"• Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"• Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"• Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"• Tienda 4: ${ingreso_tienda4:,.2f}")
print(f"→ Total combinado: ${total_general:,.2f}")
```

# **2. Ventas por categoría**
#### Calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Función para procesar cada tienda
def procesar_tienda(df, nombre_tienda):
    # Agrupar por categoría y contar las ventas
    ventas_por_categoria = df['Categoría del Producto'].value_counts().reset_index()
    ventas_por_categoria.columns = ['Categoría del Producto', 'Precio']

    # Ordenar de mayor a menor
    ventas_por_categoria = ventas_por_categoria.sort_values('Precio', ascending=False)

    return ventas_por_categoria, nombre_tienda

    # Procesar cada tienda
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}

resultados = {}
for nombre, df in tiendas.items():
    resultados[nombre] = procesar_tienda(df, nombre)[0]

# Mostrar resultados
for tiendas, df in resultados.items():
    print(f"\n{tiendas}:")
    print(df)
```

# **3. Calificación promedio de la tienda**
#### Calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
# Calcular la calificación promedio de los clientes para cada tienda
calificacion_promedio_tienda1 = tienda["Calificación"].mean()
calificacion_promedio_tienda2 = tienda2["Calificación"].mean()
calificacion_promedio_tienda3 = tienda3["Calificación"].mean()
calificacion_promedio_tienda4 = tienda4["Calificación"].mean()

# Mostrar los resultados
print("Calificacion promedio de la tienda 1:", calificacion_promedio_tienda1.round(2))
print("Calificacion promedio de la tienda 2:", calificacion_promedio_tienda2.round(2))
print("Calificacion promedio de la tienda 3:", calificacion_promedio_tienda3.round(2))
print("Calificacion promedio de la tienda 4:", calificacion_promedio_tienda4.round(2))
```

```python
# Calcular el conteo de los productos más vendidos en cada tienda
ventas_por_producto_tienda1 = tienda['Producto'].value_counts()
productos_mas_vendidos_tienda1 = ventas_por_producto_tienda1.idxmax()
ventas_tienda1 = ventas_por_producto_tienda1.max()

ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()
productos_mas_vendidos_tienda2 = ventas_por_producto_tienda2.idxmax()
ventas_tienda2 = ventas_por_producto_tienda2.max()

ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()
productos_mas_vendidos_tienda3 = ventas_por_producto_tienda3.idxmax()
ventas_tienda3 = ventas_por_producto_tienda3.max()

ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()
productos_mas_vendidos_tienda4 = ventas_por_producto_tienda4.idxmax()
ventas_tienda4 = ventas_por_producto_tienda4.max()

# Imprimir reporte de productos destacados
print("\nREPORTE DE PRODUCTOS MÁS VENDIDOS \n" +
      f"• Tienda 1: '{productos_mas_vendidos_tienda1}' (Ventas: {ventas_tienda1} unidades)\n" +
      f"• Tienda 2: '{productos_mas_vendidos_tienda2}' (Ventas: {ventas_tienda2} unidades)\n" +
      f"• Tienda 3: '{productos_mas_vendidos_tienda3}' (Ventas: {ventas_tienda3} unidades)\n" +
      f"• Tienda 4: '{productos_mas_vendidos_tienda4}' (Ventas: {ventas_tienda4} unidades)\n" +
      "\nAnálisis: Estos son los productos con mayor volumen de ventas en cada tienda")

# Calcular el conteo de los productos menos vendidos en cada tienda

ventas_por_producto_tienda1 = tienda['Producto'].value_counts()
productos_menos_vendidos_tienda1 = ventas_por_producto_tienda1.idxmin()
ventas_tienda1 = ventas_por_producto_tienda1.min()

ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()
productos_menos_vendidos_tienda2 = ventas_por_producto_tienda2.idxmin()
ventas_tienda2 = ventas_por_producto_tienda2.min()

ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()
productos_menos_vendidos_tienda3 = ventas_por_producto_tienda3.idxmin()
ventas_tienda3 = ventas_por_producto_tienda3.min()

ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()
productos_menos_vendidos_tienda4 = ventas_por_producto_tienda4.idxmin()
ventas_tienda4 = ventas_por_producto_tienda4.min()

# Imprimir reporte de productos destacados
print("\nREPORTE DE PRODUCTOS MENOS VENDIDOS\n" +
      f"• Tienda 1: '{productos_menos_vendidos_tienda1}' (Ventas: {ventas_tienda1} unidades)\n" +
      f"• Tienda 2: '{productos_menos_vendidos_tienda2}' (Ventas: {ventas_tienda2} unidades)\n" +
      f"• Tienda 3: '{productos_menos_vendidos_tienda3}' (Ventas: {ventas_tienda3} unidades)\n" +
      f"• Tienda 4: '{productos_menos_vendidos_tienda4}' (Ventas: {ventas_tienda4} unidades)\n" +
      "\nAnálisis: Estos son los productos con menor volumen de ventas en cada tienda")
```

# **5. Envío promedio por tienda**
#### Calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
# Calcular el costo de envío promedio para cada tienda
costo_envio_promedio_tienda1 = tienda["Costo de envío"].mean()
costo_envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
costo_envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
costo_envio_promedio_tienda4 = tienda4["Costo de envío"].mean()

# Mostrar los resultados
print("\nREPORTE DE COSTOS PROMEDIO DE ENVÍO POR TIENDA")
print("-----------------------------------------------")
print(f"• Tienda 1: ${costo_envio_promedio_tienda1:.2f}")
print(f"• Tienda 2: ${costo_envio_promedio_tienda2:.2f}")
print(f"• Tienda 3: ${costo_envio_promedio_tienda3:.2f}")
print(f"• Tienda 4: ${costo_envio_promedio_tienda4:.2f}")
print("-----------------------------------------------")
print(f"Promedio general: ${(costo_envio_promedio_tienda1 + costo_envio_promedio_tienda2 + costo_envio_promedio_tienda3 + costo_envio_promedio_tienda4)/4:.2f}")
```

```python
# Preparar los datos para visualización
categorias_totales = set()
datos_grafico = {}

# Obtener todas las categorías únicas y organizar los datos
for tienda_nombre, df in resultados.items():
    datos_grafico[tienda_nombre] = df.set_index('Categoría del Producto')['Precio']
    categorias_totales.update(df['Categoría del Producto'])

# Convertir a DataFrame para facilitar el plotting
df_ventas = pd.DataFrame.from_dict(datos_grafico).fillna(0)
categorias_ordenadas = df_ventas.sum(axis=1).sort_values(ascending=False).index
df_ventas = df_ventas.loc[categorias_ordenadas]

# Configurar el gráfico
plt.figure(figsize=(14, 8))
colores = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']  # Colores distintivos para cada tienda

# Crear gráfico de barras apiladas
df_ventas.plot(kind='bar', stacked=True, color=colores, edgecolor='black', linewidth=0.5)

# Personalización del gráfico
plt.title('Distribución de Ventas por Categoría en Todas las Tiendas', fontsize=16, pad=20)
plt.xlabel('Categorías de Producto', fontsize=12)
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')


# Añadir etiquetas con los totales
for i, categoria in enumerate(df_ventas.index):
    total = df_ventas.loc[categoria].sum()
    plt.text(i, total + 5, f'{int(total)}', ha='center', va='bottom')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
